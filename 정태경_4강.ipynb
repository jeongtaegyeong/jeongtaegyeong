```python
import numpy as np

x=np.linspace(0,2*np.pi,10)
y=np.sin(x)

with open("sin.txt",'w')as f:
  for i in range(len(y)):
    data = y[i]
    f.write('[%d]: %0.9f' % (i,data))
    f.write('\n')
f.close()
```
